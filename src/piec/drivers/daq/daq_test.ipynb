## 0. Setup & Connection

```python
import numpy as np
from piec.drivers.autodetect import autodetect
```

```python
# --- Option A: Auto-detect ---
daq = autodetect("daq", verbose=True)

# --- Option B: Manual ---
# from piec.drivers.daq.usb231 import USB231
# daq = USB231("USB231")
```

**PASS** if no error is raised.

### 0.1 Identification (`idn`)

```python
idn_response = daq.idn()
print("IDN:", idn_response)
```

**PASS** if identification string matches your DAQ hardware.

---
## 1. Analog Input (AI) Tests

### 1.1 Set AI Channel (`set_AI_channel`)

```python
daq.set_AI_channel(channel=0)
print("AI channel set to 0.")
```

**PASS** if no error is raised.

### 1.2 Set AI Range (`set_AI_range`)

```python
daq.set_AI_range(channel=0, range=10)
print("AI range set to ±10 V.")
```

**PASS** if no error is raised.

### 1.3 Set AI Sample Rate (`set_AI_sample_rate`)

```python
daq.set_AI_sample_rate(channel=0, sample_rate=1000)
print("AI sample rate set to 1 kHz.")
```

**PASS** if no error is raised.

### 1.4 Configure AI Channel — Combined (`configure_AI_channel`)

```python
daq.configure_AI_channel(channel=0, range=10, sample_rate=1000)
print("AI channel 0 configured: ±10 V, 1 kHz.")
```

**PASS** if no error is raised.

### 1.5 Read AI (`read_AI`)

```python
ai_value = daq.read_AI(channel=0)
print(f"AI Channel 0 reading: {ai_value}")
```

**PASS** if a reasonable voltage is returned.

### 1.6 Read AI Scan (`read_AI_scan`)

```python
scan_data = daq.read_AI_scan(channel=0, points=100, rate=1000)
print(f"AI Scan: acquired {len(scan_data)} points.")
print(f"First 5 values: {scan_data[:5]}")
```

**PASS** if 100 data points are returned.

### 1.7 Quick Read (`quick_read`)

```python
quick = daq.quick_read()
print(f"Quick Read: {quick}")
```

**PASS** if a numeric value is returned.

### 1.8 Read Data (`read_data`)

```python
data = daq.read_data(channel=0)
print(f"Read Data: {data}")
```

**PASS** if data is returned from the specified channel.

## 2. Analog Output (AO) Tests

### 2.1 Set AO Channel (`set_AO_channel`)

```python
daq.set_AO_channel(channel=0)
print("AO channel set to 0.")
```

**PASS** if no error is raised.

### 2.2 Configure AO Channel (`configure_AO_channel`)

```python
daq.configure_AO_channel(channel=0, range=10, sample_rate=1000)
print("AO channel 0 configured.")
```

**PASS** if no error is raised.

### 2.3 Write AO (`write_AO`)

```python
# Write a single voltage value
daq.write_AO(channel=0, data=[2.5])
print("Wrote 2.5 V to AO channel 0.")
```

**PASS** if measured voltage ≈ 2.5 V.

```python
daq.write_AO(channel=0, data=[0.0])
print("AO reset to 0 V.")
```

---
## 3. Digital I/O Tests

### 3.1 Digital Input (`set_DI_channel`, `read_DI`)

```python
daq.set_DI_channel(channel=0)
di_value = daq.read_DI(channel=0)
print(f"DI Channel 0: {di_value}")
```

**PASS** if a valid digital value is returned.

### 3.2 Digital Output (`set_DO_channel`, `write_DO`)

```python
daq.set_DO_channel(channel=0)
daq.write_DO(channel=0, data=[1])
print("DO Channel 0 set HIGH.")
```

**PASS** if output is HIGH.

```python
daq.write_DO(channel=0, data=[0])
print("DO Channel 0 set LOW.")
```

**PASS** if output is LOW.

### 3.3 Digital I/O Bidirectional (`set_DIO_channel`, `set_DIO_mode`)

```python
daq.set_DIO_channel(channel=0)
daq.set_DIO_mode(channel=0, mode='output')
print("DIO channel 0 set to OUTPUT mode.")
```

**PASS** if no error is raised.

```python
daq.set_DIO_mode(channel=0, mode='input')
print("DIO channel 0 set to INPUT mode.")
```

**PASS** if no error is raised.

### 3.4 Configure DIO (`configure_DIO_channel`)

```python
daq.configure_DIO_channel(channel=0, mode='input', sample_rate=1000)
print("DIO configured: input, 1 kHz.")
```

**PASS** if no error is raised.
