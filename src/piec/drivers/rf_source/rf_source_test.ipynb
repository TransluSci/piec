# RF Source Driver Test Notebook

| Layer | Class | Key Methods |
|---|---|---|
| 1 | `Instrument` | `idn()` |
| 2 | `Scpi` | `reset()`, `clear()`, `error()`, `self_test()`, `operation_complete()`, `initialize()` |
| 3 | `RF_source` | Frequency, power, modulation (AM/FM/PM), sweep |

**Instructions:** Run each cell sequentially. A spectrum analyzer or power meter is recommended for verification.

**Safety Note:** RF sources can output significant power. Ensure proper termination (50 Ω) on the output.

---
## 0. Setup & Connection

```python
from piec.drivers.autodetect import autodetect
```

```python
# --- Option A: Auto-detect ---
rf = autodetect("rf_source", verbose=True)

# --- Option B: Manual ---
# from piec.drivers.rf_source.<driver> import <DriverClass>
# rf = <DriverClass>("GPIB0::20::INSTR")
```

**PASS** if no error is raised.

---
## 1. Instrument & SCPI Tests

```python
idn_response = rf.idn()
print("IDN:", idn_response)
```

**PASS** if identification string matches your instrument.

```python
rf.reset()
print("Reset: OK")

rf.clear()
print("Clear: OK")

print("Error:", rf.error())
print("Self-test:", rf.self_test())
print("OPC:", rf.operation_complete())

rf.initialize()
print("Initialize: OK")
```

**Expected:** Error: `0`, Self-test: `0`, OPC: `1`

**PASS** if all values match.

---
## 2. Core RF Output Tests

### 2.1 Set Frequency (`set_frequency`)

```python
rf.set_frequency(frequency=1e9)
print("Frequency set to 1 GHz.")
```

**PASS** if frequency matches.

```python
rf.set_frequency(frequency=100e6)
print("Frequency set to 100 MHz.")
```

**PASS** if frequency reads 100 MHz.

### 2.2 Set Power (`set_power`)

```python
rf.set_power(power=-10)
print("Power set to -10 dBm.")
```

**PASS** if power matches.

```python
rf.set_power(power=0)
print("Power set to 0 dBm.")
```

**PASS** if power reads 0 dBm.

### 2.3 Output On/Off (`output`)

```python
rf.output(channel=1, on=True)
print("RF output ON.")
```

**PASS** if output is ON.

```python
rf.output(channel=1, on=False)
print("RF output OFF.")
```

**PASS** if output turns OFF.

## 3. Modulation Tests

### 3.1 Set Modulation Type (`set_modulation`)

```python
rf.set_modulation(modulation_type='AM')
print("Modulation type set to AM.")
```

**PASS** if modulation type shows AM.

### 3.2 Enable/Disable Modulation (`enable_modulation`)

```python
rf.enable_modulation(on=True)
print("Modulation ON.")
```

**PASS** if modulation indicator shows active.

```python
rf.enable_modulation(on=False)
print("Modulation OFF.")
```

**PASS** if modulation is disabled.

### 3.3 AM Parameters

```python
rf.set_modulation(modulation_type='AM')
rf.set_am_depth(depth=50)
print("AM depth set to 50%.")
```

**PASS** if AM depth reads 50%.

**PASS** if AM modulation frequency reads 1 kHz.

**PASS** if FM mod frequency reads 1 kHz.
