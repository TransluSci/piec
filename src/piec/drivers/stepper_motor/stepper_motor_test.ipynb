---
## 0. Setup & Connection

```python
from piec.drivers.autodetect import autodetect
```

```python
# --- Option A: Auto-detect ---
stepper = autodetect("stepper_motor", verbose=True)

# --- Option B: Manual ---
# from piec.drivers.stepper_motor.arduino_stepper import ArduinoStepper
# stepper = ArduinoStepper("COM3")  # Replace with actual serial port
```

**PASS** if no error is raised.

---
## 1. Identification (`idn`)

```python
idn_response = stepper.idn()
print("IDN:", idn_response)
```

**PASS** if an identification string is returned.

---
## 2. Step Tests (`step`)

### 2.1 Step Forward

```python
# Step 100 steps in the forward direction
stepper.step(steps=100, dir=1)
print("Stepped 100 steps forward.")
```

**PASS** if the motor rotates forward.

### 2.2 Step Backward

```python
# Step 100 steps in the reverse direction
stepper.step(steps=100, dir=-1)
print("Stepped 100 steps backward.")
```

**PASS** if the motor rotates backward.

### 2.3 Small Step

```python
# Step 1 step forward
stepper.step(steps=1, dir=1)
print("Stepped 1 step forward.")
```

**PASS** if the motor makes a tiny movement.

### 2.4 Large Step

```python
# Step 1000 steps forward
stepper.step(steps=1000, dir=1)
print("Stepped 1000 steps forward.")
```

**PASS** if the motor rotates a significant amount.

```python
# Return to original position
stepper.step(steps=1000, dir=-1)
print("Stepped 1000 steps backward (return).")
```
