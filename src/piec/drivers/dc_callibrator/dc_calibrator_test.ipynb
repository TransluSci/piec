# DC Calibrator Driver Test Notebook

| Layer | Class | Key Methods |
|---|---|---|
| 1 | `Instrument` | `idn()` |
| 2 | `DCCalibrator` | `output()`, `set_output()`, `set_voltage()`, `set_current()`, `reset()`, `error()` |

**Instructions:** Run each cell sequentially. A DMM connected to the calibrator output is recommended to verify output values.

⚠️ **Safety Note:** DC Calibrators can source precise voltages and currents. Ensure your measurement device can handle the configured values.

---
## 0. Setup & Connection

```python
from piec.drivers.autodetect import autodetect
```

```python
# --- Option A: Auto-detect ---
cal = autodetect("dc_callibrator", verbose=True)

# --- Option B: Manual ---
# from piec.drivers.dc_callibrator.edc522 import EDC522
# cal = EDC522("GPIB0::5::INSTR")
```

✅ **PASS** if no error is raised.

---
## 1. Identification (`idn`)

```python
idn_response = cal.idn()
print("IDN Response:")
print(idn_response)
```

✅ **PASS** if the string matches your physical instrument.

---
## 2. Reset (`reset`)

```python
cal.reset()
print("Reset command sent.")
```

**Expected Result:** Calibrator returns to a safe state (output crowbarred/disabled).

✅ **PASS** if the instrument resets.

---
## 3. Error Query (`error`)

```python
error_response = cal.error()
print("Error:", error_response)
```

**Expected Result:** No errors (returns `0` or equivalent clean status).

✅ **PASS** if no errors reported.

---
## 4. Output Control (`output`)

```python
cal.output(on=True)
print("Output ON.")
```

**Expected Result:** Output is **enabled**. The calibrator's output indicator should show ON (un-crowbarred).

✅ **PASS** if output indicator shows ON.

```python
cal.output(on=False)
print("Output OFF.")
```

**Expected Result:** Output is **disabled** (crowbarred). No voltage/current at terminals.

✅ **PASS** if output indicator shows OFF.

---
## 5. Set Voltage (`set_voltage`)

```python
cal.set_voltage(voltage=1.0)
cal.output(on=True)
print("Voltage set to 1.0 V. Output ON.")
```

**Expected Result:** The calibrator should output **1.000 V**. Verify with a DMM on the output terminals.

✅ **PASS** if measured voltage ≈ 1.0 V.

```python
cal.set_voltage(voltage=5.0)
print("Voltage set to 5.0 V.")
```

✅ **PASS** if measured voltage ≈ 5.0 V.

```python
cal.set_voltage(voltage=0.0)
print("Voltage set to 0 V.")
```

✅ **PASS** if measured voltage ≈ 0 V.

---
## 6. Set Current (`set_current`)

```python
cal.set_current(current=0.001)
print("Current set to 1 mA.")
```

**Expected Result:** The calibrator should output **1 mA**. Verify with a DMM in current mode.

✅ **PASS** if measured current ≈ 1 mA.

```python
cal.set_current(current=0.0)
print("Current set to 0 A.")
```

✅ **PASS** if measured current ≈ 0 A.

---
## 7. Set Output — Generic (`set_output`)

```python
cal.set_output(value=2.5, mode='voltage')
print("Output set to 2.5 V via set_output().")
```

✅ **PASS** if measured voltage ≈ 2.5 V.

```python
cal.set_output(value=0.005, mode='current')
print("Output set to 5 mA via set_output().")
```

✅ **PASS** if measured current ≈ 5 mA.

---
## 8. Cleanup

```python
cal.output(on=False)
cal.reset()
print("Output OFF. Calibrator reset.")
```
