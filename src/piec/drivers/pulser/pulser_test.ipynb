# Pulser Driver Test Notebook

| Layer | Class | Key Methods |
|---|---|---|
| 1 | `Instrument` | `idn()` |
| 2 | `Scpi` | `reset()`, `clear()`, `error()`, `self_test()`, `operation_complete()`, `initialize()` |
| 3 | `Pulser` | Timing, levels, trigger/burst, output |

**Instructions:** Run each cell sequentially. An oscilloscope on Channel 1 output is recommended.

**Safety Note:** Ensure the pulser output is connected to a proper load (50 Ω).

---
## 0. Setup & Connection

```python
from piec.drivers.autodetect import autodetect
```

```python
# --- Option A: Auto-detect ---
pulser = autodetect("pulser", verbose=True)

# --- Option B: Manual ---
# from piec.drivers.pulser.bnc765 import BNC765
# pulser = BNC765("GPIB0::15::INSTR")
```

**PASS** if no error is raised.

---
## 1. Instrument & SCPI Tests

```python
idn_response = pulser.idn()
print("IDN:", idn_response)
```

**PASS** if `Manufacturer, Model, SN, FW` string matches your instrument.

```python
pulser.reset()
print("Reset: OK")

pulser.clear()
print("Clear: OK")

print("Error:", pulser.error())
print("Self-test:", pulser.self_test())
print("OPC:", pulser.operation_complete())

pulser.initialize()
print("Initialize: OK")
```

**Expected:** Error: `0`, Self-test: `0`, OPC: `1`

**PASS** if all values match.

---
## 2. Pulse Timing Tests

### 2.1 Set Period (`set_period`)

```python
pulser.set_period(channel=1, period=0.001)
print("Period set to 1 ms (1 kHz).")
```

**PASS** if period matches.

### 2.2 Set Frequency (`set_frequency`)

```python
pulser.set_frequency(channel=1, frequency=10000)
print("Frequency set to 10 kHz.")
```

**PASS** if frequency reads 10 kHz.

### 2.3 Set Width (`set_width`)

```python
pulser.set_width(channel=1, width=10e-6)
print("Pulse width set to 10 µs.")
```

**PASS** if width matches.

### 2.4 Set Delay (`set_delay`)

```python
pulser.set_delay(channel=1, delay=5e-6)
print("Delay set to 5 µs.")
```

**PASS** if delay matches.

### 2.5 Set Rise / Fall Time

```python
pulser.set_rise_time(channel=1, rise_time=5e-9)
print("Rise time set to 5 ns.")
```

**PASS** if rise time matches or is set to the nearest available value.

```python
pulser.set_fall_time(channel=1, fall_time=5e-9)
print("Fall time set to 5 ns.")
```

**PASS** if fall time matches.

## 5. Output Test

```python
# Set to continuous mode for output test
pulser.set_trigger_mode(mode='CONT')
pulser.set_trigger_source(source='INT')
pulser.set_frequency(channel=1, frequency=1000)
pulser.set_high_level(channel=1, high_level=3.3)
pulser.set_low_level(channel=1, low_level=0.0)
pulser.set_width(channel=1, width=100e-6)

pulser.output(channel=1, on=True)
print("Output ON — verify on oscilloscope.")
```

**PASS** if the pulse waveform is visible and correct.

```python
pulser.output(channel=1, on=False)
print("Output OFF.")
```

**PASS** if output stops.

---
## 6. Cleanup

```python
pulser.output(channel=1, on=False)
pulser.reset()
print("Pulser reset.")
```
